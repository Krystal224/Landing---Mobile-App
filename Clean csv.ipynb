```python
import csv
import json
import pandas as pd

csvFilePath = "articles0.csv"
jsonFilePath = "articles0.json"

```

```python
data = {}
with open(csvFilePath) as csvFile:
    csvReader = csv.DictReader(csvFile)
    for csvRow in csvReader:
        author = csvRow["author"]
        data[author] = csvRow

with open(jsonFilePath, "w") as jsonFile:
    jsonFile.write(json.dumps(data, indent = 4))
```

```python
# print(len(df0.claps[0]))
# t = df0.claps[0]


## Get all the rows with K out to df_K
df_K = pd.DataFrame()

for i in range(df0.shape[0]):
    each_r = df0.claps[i]
    for j in range(len(each_r)):
        if each_r[j] == 'K':
            df_K = df_K.append(df0.iloc[i:(i+1)])
            

# df0.claps[0] = df0.claps[0] + 'K'
# df0.claps[0]

df_K.index = range(196)
df_K.head(5)
```

```python
# Remove K from the claps column, convert string to number and sort the column from high to low

# function to remove_K
def remove_K(string):
    string = string.replace("K","")
    return string
  
# a = remove_K(df_K.claps[0])
# df_K.claps[0] = a
# df_K.claps[0]

# a = df_K.claps[11]
# a




# Remove K for all claps   
for i in range(df_K.shape[0]):
    each_str = df_K.claps[i]
    each_str = each_str.replace("K","")
    df_K.claps[i] = each_str
    


# Sort the column by claps after Removing 'K'
   
df_K['claps'] = pd.to_numeric(df_K.claps, errors='coerce')#convert to numerical value for sorting purposes
df_K['claps'] = df_K['claps'].astype('int')
df_K = df_K.sort_values(by='claps', ascending=False)


# Add 'K' to all claps
df_K.index = range(196)

df_K['claps'] = df_K['claps'].apply(str)

for j in range(df_K.shape[0]):
    a = df_K.claps[j] + 'K'
    df_K.claps[j] = a
  

```
